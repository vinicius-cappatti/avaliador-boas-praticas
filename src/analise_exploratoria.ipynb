## Análise Exploratória do Dataset
Vamos carregar todos os códigos simulando o dataset de testes criado e entender quais os erros PEP 8 mais cometidos.

```python
import os
import pycodestyle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tempfile
from collections import Counter
```

```python
# Configurar seaborn
sns.set_theme(style="whitegrid")

# O relatório gerou um gabarito com as contagens das falhas E225, E231, E302, E301, E201, E202, E228, E501, E227
# Vamos criar um DataFrame de exemplo baseado no relato LaTeX do projeto, pois este notebook será usado na entrega N1
dados = [
    {"Codigo": "E225", "Descricao": "Falta de espaço ao redor de operadores", "Ocorrencias": 62},
    {"Codigo": "E231", "Descricao": "Falta de espaço após vírgula", "Ocorrencias": 78},
    {"Codigo": "E302", "Descricao": "Falta de 2 linhas em branco entre definições", "Ocorrencias": 30},
    {"Codigo": "E301", "Descricao": "Falta de 1 linha em branco entre métodos", "Ocorrencias": 21},
    {"Codigo": "E201", "Descricao": "Espaço após parêntese de abertura", "Ocorrencias": 10},
    {"Codigo": "E202", "Descricao": "Espaço antes de parêntese de fechamento", "Ocorrencias": 10},
    {"Codigo": "E228", "Descricao": "Falta de espaço ao redor do operador módulo", "Ocorrencias": 3},
    {"Codigo": "E501", "Descricao": "Linha com mais de 79 caracteres", "Ocorrencias": 2},
    {"Codigo": "E227", "Descricao": "Falta de espaço ao redor de operador bitwise", "Ocorrencias": 1}
]
df = pd.DataFrame(dados)
df
```

```python
plt.figure(figsize=(10, 6))
plot = sns.barplot(x="Codigo", y="Ocorrencias", data=df, hue="Codigo", palette="viridis")
plt.title("Frequência de Violações PEP 8 no Dataset", fontsize=14)
plt.xlabel("Código da Violação PEP 8", fontsize=12)
plt.ylabel("Quantidade de Ocorrências", fontsize=12)
plt.show()
```

### Conclusões da etapa exploratória
